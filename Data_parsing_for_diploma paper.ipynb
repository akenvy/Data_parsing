```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import pandas as pd
```

```python
pd.options.display.max_columns = 450
pd.options.display.max_rows = 150
```

```python
dfr_list = []
for year in range(41):
    url_1 = 'https://www.formula1points.com/season/season-progression/'
    url_2 = 1980
    url = url_1 + str(url_2 + year)
    response = requests.get(url, headers={'User-Agent': UserAgent().chrome})
    soup = BeautifulSoup(response.text)
    
    columns = []
    columns.append(soup.thead.find_all('th')[1].text)
    for i in list(soup.thead.find_all('th')[2:]):
        columns.append(i.text+'_'+url[-4:])
    df = pd.DataFrame(columns = columns)

    for j in range(2, len(soup.tbody)):
        a = []
        a.append(list(soup.tbody)[j].a.text)
        for i, obj in enumerate(list(soup.tbody)[j].find_all('td')):
            a.append(obj.text)
        df.loc[j-2] = a
    dfr_list.append(df.Driver.values)
```
